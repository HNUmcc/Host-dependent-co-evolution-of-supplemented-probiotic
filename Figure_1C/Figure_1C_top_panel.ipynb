## Figure 1B. The temporal dynamics of # of probiotic SNPs in all three hosts.

```r
library('ggplot2')

outpath='./'
```

```r
a<-read.table("snp_number.txt", sep='\t', header=TRUE)
head(a)
```

```r
p<-ggplot(a, aes(x=as.factor(TIME.DAY.), y=SNP))+
  xlab("Time (days)")+
  ylab("# of SNP detected")+
  geom_boxplot(alpha=0.4, outlier.shape = NA) + 
  scale_color_manual(values=c("#FFF68F","#00BFFF", "#C1FFC1"))+
  geom_jitter(aes(color=MODEL), position=position_jitterdodge(jitter.width=0.2, dodge.width = 0.8),size=3) + #
  theme_bw()+
  theme(#axis.text.x = element_text(angle = 90, vjust=0.5, hjust = 1,size=12),
        axis.text.y = element_text(size=12),
        axis.text.x = element_text(size=12), 
        axis.title = element_text(size=18),
        axis.line = element_line(color="black"),
        strip.background = element_rect(colour = "white"),
        panel.border = element_blank())
p
ggsave(filename=paste(outpath, "Num_of_snp.boxplot.pdf",sep=""),plot=p, height=4, width=5)

```
